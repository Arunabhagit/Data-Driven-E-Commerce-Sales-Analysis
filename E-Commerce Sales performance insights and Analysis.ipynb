```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv('/content/drive/MyDrive/Data Analytics Project/E-Commerce_Store_Data.csv',encoding = 'latin-1')
```

```python
data
```

#Data Cleaning

#Remove Duplicate Data

```python
duplicates = data.duplicated()
data[duplicates]
```

```python
data.drop_duplicates(inplace=True)
```

```python
duplicates = data.duplicated()
data[duplicates]
```

#Replace Missing Data

```python
missing_data = data.isnull().sum()
print("Missing Value:")
missing_data
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy ='most_frequent')
imputer.fit(data[['Quantity']])
data[['Quantity']] = imputer.transform(data[['Quantity']])
```

```python
imputer = SimpleImputer(strategy ='most_frequent')
imputer.fit(data[['Country']])
data[['Country']] = imputer.transform(data[['Country']])
```

```python
imputer = SimpleImputer(strategy ='mean')
imputer.fit(data[['Sales']])
data[['Sales']] = imputer.transform(data[['Sales']])
```

```python
missing_data = data.isnull().sum()
print("Missing Value:")
missing_data
```

#Assing Correct Datatype

```python
data.info()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

#Create New Column of Year,Month and Day

```python
data['Order Month'] = data['Order Date'].dt.strftime('%B')
data['Order Year'] = data['Order Date'].dt.year
data['Order Day by week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

```python
data['date_difference'] = (data['Ship Date'] - data['Order Date'])
data
```

#Data Inshights and Analysis

#Analyze Yearly Sales and Profit trends

```python
yearly_sales_profit = data.groupby('Order Year').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
yearly_trends = round(yearly_sales_profit)
yearly_trends
```

```python
color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=yearly_trends['Order Year'],
                     y=yearly_trends['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=yearly_trends['Order Year'],
                     y=yearly_trends['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Yearly Sales and Profit Trends',
                  xaxis_title='Customer Segment', yaxis_title='Amount')

fig.show()
```

#Analyze the Monthly sales of the store and identify trends of sales in every month

```python
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

data['Order Month'] = pd.Categorical(data['Order Month'], categories=month_order, ordered=True)



sales_months = data.groupby('Order Month') ['Sales'].sum().reset_index()

sales_by_month = sales_months.sort_values('Order Month').reset_index(drop=True)

sales_by_months = round(sales_by_month,2)

sales_by_months
```

```python
fig = px.line(sales_by_months,x='Order Month',y = 'Sales' , title = 'Month Sales Trends')
fig.show()
```

#Analyze Monthly profits Trend according to Sales

```python

month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

data['Order Month'] = pd.Categorical(data['Order Month'], categories=month_order, ordered=True)
profit = data.groupby('Order Month')['Profit'].sum().reset_index()
profit_by_month = round(profit)
profit_by_month
```

```python
fig = px.line(profit_by_month ,x='Order Month',y = 'Profit' , title = 'Monthly Profit Trends')
fig.show()
```
